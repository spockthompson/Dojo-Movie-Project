```python
#Data Sets
akas_url="https://datasets.imdbws.com/title.akas.tsv.gz"
basics_url="https://datasets.imdbws.com/title.basics.tsv.gz"
ratings_url="https://datasets.imdbws.com/title.ratings.tsv.gz"
```

```python
import pandas as pd
import numpy as np
```

```python
basics = pd.read_csv(basics_url, sep='\t', low_memory=False)

akas = pd.read_csv(basics_url, sep='\t', low_memory=False)

ratings = pd.read_csv(basics_url, sep='\t', low_memory=False)
```

```python
basics.head(10)
```

```python
basics.info()
```

```python
basics.replace({'\\N':np.nan}, inplace = True)
```

```python
# Eliminate movies that are null for runtimeMinutes
basics.dropna(subset=['runtimeMinutes'], inplace = True)
basics.dropna(subset=['genres'], inplace = True)
```

```python
basics['titleType'].value_counts()
```

```python
mask1 = (basics['titleType'] == 'movie')
filter_basics = basics.loc[mask1]
```

```python
mask2 = (basics['startYear'] >= '2000') & (basics['startYear'] < '2022')
filter_basics = filter_basics.loc[mask2]
```

```python
filter_basics.head()
```

```python
# Exclude movies that are included in the documentary category.
is_documentary = filter_basics['genres'].str.contains('documentary',case=False)
filter_basics = filter_basics[~is_documentary]
```
